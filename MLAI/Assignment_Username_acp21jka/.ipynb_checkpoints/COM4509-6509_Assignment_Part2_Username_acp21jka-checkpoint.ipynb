```python
def set_seed(seed):
    # 1. Set `PYTHONHASHSEED` environment variable at a fixed value
    os.environ["PYTHONHASHSEED"] = str(seed)
    # 2. Set `python` built-in pseudo-random generator at a fixed value
    random.seed(seed)
    # 3. Set `numpy` pseudo-random generator at a fixed value
    np.random.seed(seed)
    # 4. Set `pytorch` pseudo-random generator at a fixed value
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False
        
set_seed(seed=66505)
```

## B2- Logistic regression
### B2(a)- SKLearn model

We train the SKLearn model on 5 different hyperparameters(l2 regularisation parameter) and check the RMSE over val set

```python
start1=time.time()
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error

for l2_param in np.logspace(start=-1,stop=-5,num=5):
    
    LR=LogisticRegression(penalty='l2',C=l2_param)
    
    LR.fit(reshape(Xtrain),np.ravel(ytrain)) # to make the data as a 1D column, we use np.ravel 
    
    ypred=LR.predict(reshape(Xval))
    
    print('For regularisation parameter={:.5f}, RMSE over validation data is {:.3f}' 
          .format(l2_param, np.sqrt(mean_squared_error(np.ravel(yval), ypred))))
```

We select the hyperparameters from above and train the model on the set hyperparameters. It is rp=0.1 in this case

```python
LR_final=LogisticRegression(penalty='l2',C=0.1)  #l2 param = 0.1 chosen because it produces lowest rmse over validation set

# We combine train and val data to train the model using best hyperparameter
X_train_and_val=torch.cat((Xtrain,Xval))      
y_train_and_val=torch.cat((ytrain,yval))

LR_final.fit(reshape(X_train_and_val),np.ravel(y_train_and_val))  # We fit the model now

y_predicted=LR_final.predict((reshape(Xtest)))  # Predicting values over Test dataset

RMSE_A=np.sqrt(mean_squared_error(np.ravel(ytest), y_predicted)) # Calculating RMSE over test set

print('RMSE error of SKLearn Logistic Regression trained model on test set is:{:.3f}'.format(RMSE_A))

# We calculate the trained accuraccy on the three datasets(training,validation,test) 
# using model.score and label them M1_a,M2_a and M3_a

M1_a=LR_final.score( reshape(Xtrain) , np.ravel(ytrain) )*100
M2_a=LR_final.score( reshape(Xval)   , np.ravel(yval) )*100
M3_a=LR_final.score( reshape(Xtest)  , np.ravel(ytest) )*100
print('The accuracy percent of each metric from SKLearn Logistic Regression model is M1={:.2f} %, M2={:.2f} %, M3={:.2f} %'
      .format(M1_a,M2_a,M3_a))
end1=time.time()
```

### B2(b)- PyTorch model

We create the logistic regression module using PyTorch

```python
start2=time.time()
class LogisticRegression(torch.nn.Module):
    def __init__(self):
        super(LogisticRegression, self).__init__()
        self.linear = torch.nn.Linear(28*28, 1)     #The input is an image with dimension 28*28. We ouput 
                                                    # single neuron which contains the probability between 0 and 1
    def forward(self, x):
            y_pred = torch.sigmoid(self.linear(x)) # If probability>0.5, the neuron suggests class 1, else class 0
            return y_pred
        
def parameter_reset(model):                # We reset parameter weights, just so that previous result won't affect the next iterations
    if isinstance(model, nn.Conv2d) or isinstance(model, nn.Linear):
        model.reset_parameters()
```

We use the gradient update formula with l2 regularisation for our model training

w = w - (lr) x ( dE/dw + (lambda * w / N) )

w=weights,
lr= learning rate,
N= batch size,
dE/dw= gradient of loss function w.r.t w,
lambda= l2 regularization parameter

```python
# We define a criteria for the loss measurement. It is Mean Squared Error here
criterion = torch.nn.MSELoss() 

# We train the above LR model for various hyperparameters like lr and rp, each for 30 epochs, and later check loss on the validation set
# Hyperparameters are learning rate and the l2 regularization coefficient, and then we choose best one for further training

matrix_of_losses=np.empty(shape=(12,3))                         # We make a loss matrix containing losses for each grid value of lr,rp
p=0                                                             # Indexing matrix
for l2_reg in np.logspace(start=-4,stop=-1,num=4):              # for regularization coefficient
    for lr in [0.3,0.35,0.4]:                                   # for learning rate
        model=LogisticRegression()
        model.apply(parameter_reset) 
        for epoch in range(30):                                 # Train on 30 epochs initially
            for x,y in train_loader:
                model.zero_grad()                               # Reset gradients
                x=reshape(x)                                    # reshape from n*1*28*28 to n*784
                N=x.shape[0]                                    # no. of samples in the batch
                images=x.to(torch.float32)                      # convert to float
                labels=y.to(torch.float32)
                outputs = model(images)                         # forward pass
                loss =criterion(outputs, labels)                # the loss to be minimised
                loss.backward()                                 # Backward pass for autograd
                
                for W in model.parameters():                    # we apply gradient update, we update the gradient manually
                    W.data-= lr*(W.grad.data+(l2_reg*W.data/N)) 
                    
        with torch.no_grad(): #testing phase, no need to compute the gradients to save time
            running_loss=0
            no_of_batches=0
            for batch_data in val_loader_at_eval:
                img,lab=batch_data
                img=reshape(img)
                img=img.to(torch.float32)
                lab=lab.to(torch.float32)
                pred=model(img)
                loss=criterion(pred,lab)
                running_loss=running_loss + loss.item() # add all losses to calculate average loss at last
                no_of_batches+=1
            matrix_of_losses[p,:]=running_loss,l2_reg,lr
            p+=1
        print('Running loss= {:.3f} on validation data for regularization parameter={} and learning rate={}'
              .format(running_loss/no_of_batches,l2_reg,lr))       
```

```python
#finding lowest loss and corresponding learning rate and regularisation parameters from the matrix
best_params=matrix_of_losses[np.argmin(matrix_of_losses,axis=0)[0]] 
print('lowest loss on val set={:.3f},l2_param={},lr={}'.format(best_params[0],best_params[1],best_params[2]))
```

As per the tests above, we choose the best hyperparameters, and train our model on those hyperparameters. The best hyperparameters are lr=0.3 and l2_reg=0.01

```python
best_LR_model = LogisticRegression()
best_LR_model.apply(parameter_reset)       # reset the parameters for further exploration
epochs, avg_losses=[], []

for epoch in range(31): #train the model on best parameters further for 31 epochs with chosen hyperparameters
    running_loss=0
    batch_number=0
    lr     = best_params[2]   #chosen from the above matrix
    l2_reg = best_params[1]
    for loader in [val_loader,train_loader]:
        for x,y in loader:
            x=reshape(x)                                              # reshape our input from N*1*28*28 to N*784
            best_LR_model.zero_grad()                                    # reset the gradient
            N=x.shape[0]                                              # number of data i.e batch size
            # convert data to torch float
            images, labels =x.to(torch.float32), y.to(torch.float32)
            outputs = best_LR_model(images)                              # forward pass
            loss =criterion(outputs, labels)                          # calculate loss
            loss.backward()                                           # backward pass
            
            # we update the gradient manually using w formula as written above
            for param in best_LR_model.parameters():
                param.data-= lr*(param.grad.data+(l2_reg*param.data/N))
            
            batch_number+=1
            running_loss =running_loss + loss.item()
    if (epoch % 2) == 0:         # We track the loss for every 2 epochs 
        avg_losses.append(float(running_loss/no_of_batches))
        epochs.append(epoch+1)
    if (epoch % 10) == 0:
        print('After epoch-{}, average loss is {:.4f}'.format(epoch+1, float(running_loss/batch_number)))
```

```python
plt.figure(figsize=(6, 4))
plt.plot(epochs,avg_losses)
plt.title('Average loss against number of epoch')
plt.xlabel('Number of Epochs')
plt.ylabel('Average loss')
plt.show()
```

```python
yPredicted=[] # List to store predicted values
for i in range(ytest.shape[0]):
    val=best_LR_model(reshape(Xtest)[i]) # Calculating the probability output from single neuron
    if val<=0.5:
        yPredicted.append(0) # If prob<0.5, it means class 0, else it is class 1
    else:
        yPredicted.append(1)
RMSE_B=np.sqrt(mean_squared_error(np.ravel(ytest), yPredicted)) # we calculate the RMSE over test set with the new model
print('RMSE error of trained Logistic Regression model on test set is: {:.3f}'.format(RMSE_B))
```

We define a function to calculate the accuracy score of datasets using the above trained model

```python
def calc_accu_score(dataloader):
    with torch.no_grad():
        for x,y in dataloader:
            x_dataset=x
            y_dataset=y
        val_out=[]
        for i in range(0,y_dataset.shape[0]):
            val=best_LR_model(reshape(x_dataset)[i])
            if val<=0.5:
                val_out.append(0)
            else:
                val_out.append(1)
        correct=0
        for i in range(0,len(val_out)):
            if val_out[i]==y_dataset[i]:
                correct+=1
        return (correct*100/len(val_out))
```

But it does seem that the accuracy on test set from PyTorch model is a nearly equal in comparison to the inbuilt one. 

#### 2.
We used only 1 linear layer which takes in 784 weights and baises and outputs 1 neuron after sigmoid(Logit) operation. We could optimise the network by using 2 or more such linear layers, and also design a network such that it outputs 2 neurons where our neurons output the probability for the two available classes and the one with highest probability is picked as the output predicted by the network. In this notebook however, we have chosen to do it on 1 Linear layer and 1 output neuron, and still we got acceptable results, even better than those we have gotten from SKLearn model.

#### 3.
Although the SkLearn Logistic Regression has low accuracy percent in comparison to PyTorch Logistic Regression, I would prefer the inbuilt SkLearn Logistic Regression given how fast it can train the model(0.5 seconds for testing over valset, then training, then calculating accuracy, while PyTorch model took around 45 seconds to do the same set of operations). However, in some fields like Medicine, it would be better to train a PyTorch model from scratch to get the best accuracy possible, as even small percentage of False Negative results will impact the society very badly.

## C2- Convolutional neural networks
### C2(a) Two Conv layers and two FC layers

```python
# We convert our data to torch.cuda tensor types if cuda is available
if torch.cuda.is_available():
    device=torch.device('cuda')
else:
    device=torch.device('cpu')

# We define our loss criteria. This criterion computes the cross entropy loss between input and the target.
criterion = nn.CrossEntropyLoss()

# We reset parameter weights so that previous result won't affect next iteration i.e. we do model.reset_params
def parameter_weight_reset(model):
    with torch.no_grad():
        reset_param = getattr(model,"reset_parameters", None) #If attribute is callable,execute model.reset_parameters
        if callable(reset_param):
            model.reset_parameters()
```

We now create 4 classes and 4 objects each having Two Conv layers and two FC layers

```python
class CNN_A_1(nn.Module):
    def __init__(self):
        super(CNN_A_1, self).__init__()
        self.conv1 = nn.Conv2d(1, 3, 5,padding=2) #1: #input channels; 3: #output channels; 5: kernel size; 2: padding
        self.conv2 = nn.Conv2d(3, 5, 7,padding=1)
        self.fc1 = nn.Linear(5 * 11 * 11, 50)
        self.fc2 = nn.Linear(50, 4)
        self.pool1 = nn.MaxPool2d(3, 1)
        self.pool2 = nn.MaxPool2d(2, 2)

    def forward(self, x):
        x = self.pool1(torch.sigmoid(self.conv1(x))) # We use torch.sigmoid as our activation function in all layers here
        x = self.pool2(torch.sigmoid(self.conv2(x)))
        x = x.view(-1, 5 * 11 * 11)
        x = torch.sigmoid(self.fc1(x))
        x = torch.sigmoid(self.fc2(x))
        return x
        
myCNN_A_1 = CNN_A_1().to(device)          # convert to CUDA if available
summary(myCNN_A_1, input_size=(1,28,28))  # summary of the model created above
```

```python
class CNN_A_2(nn.Module):
    def __init__(self):
        super(CNN_A_2, self).__init__()
        self.conv1 = nn.Conv2d(1, 8, 6,padding=2) #1: #input channels; 8: #output channels; 6: kernel size; 2: padding
        self.conv2 = nn.Conv2d(8, 32, 3)
        self.pool1 = nn.MaxPool2d(3, dilation=2)
        self.fc1 = nn.Linear(32 * 6 * 6, 30)
        self.fc2 = nn.Linear(30, 4)

    def forward(self, x):
        x = self.pool1(F.relu(self.conv1(x)))
        x = F.relu(self.conv2(x))
        x = x.view(-1, 32 * 6 * 6)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
        
myCNN_A_2= CNN_A_2().to(device)          # convert to CUDA if available
summary(myCNN_A_2, input_size=(1,28,28))  # summary of the model created above
```

```python
class CNN_A_3(nn.Module):
    def __init__(self):
        super(CNN_A_3, self).__init__()
        self.conv1 = nn.Conv2d(1, 3, 5,padding=2) #1: #input channels; 3: #output channels; 5: kernel size; 2: padding
        self.conv2 = nn.Conv2d(3, 5, 7,padding=1)
        self.fc1 = nn.Linear(5 * 11 * 11, 50)
        self.fc2 = nn.Linear(50, 4)
        self.pool1 = nn.MaxPool2d(3, 1)
        self.pool2 = nn.MaxPool2d(2, 2)

    def forward(self, x):
        x = self.pool1(F.relu(self.conv1(x)))  # We use F.relu as our activation function for all layers in this model
        x = self.pool2(F.relu(self.conv2(x)))
        x = x.view(-1, 5 * 11 * 11)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        return x
        
myCNN_A_3 = CNN_A_3().to(device)          # convert to CUDA if available
summary(myCNN_A_3, input_size=(1,28,28))  # summary of the model created above
```

```python
class CNN_A_4(nn.Module):
    def __init__(self):
        super(CNN_A_4, self).__init__()
        self.conv1 = nn.Conv2d(1, 16, 5,padding=2) #1: #input channel; 16: #output channels; 5: kernel size; 2: padding
        self.conv2 = nn.Conv2d(16, 50, 3,padding=1)
        self.pool1 = nn.MaxPool2d(2,2)
        self.fc1 = nn.Linear(50 * 7 * 7, 30)
        self.fc2 = nn.Linear(30, 4)

    def forward(self, x):
        x = self.pool1(F.relu(self.conv1(x))) # We apply 2x2 pooling with stride=2, on both the conv layers
        x = self.pool1(F.relu(self.conv2(x)))
        x = x.view(-1, 50 * 7 * 7)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
        
myCNN_A_4= CNN_A_4().to(device)          # convert to CUDA if available
summary(myCNN_A_4, input_size=(1,28,28))  # summary of the model created above
```

```python
loss_dict=dict()          # We create a dictionary to store the model name and the losses on validation set using that model
lr=0.01
momentum=0.9
for model in [myCNN_A_1, myCNN_A_2, myCNN_A_3, myCNN_A_4]:           # Run the loop for each model
    model.apply(parameter_weight_reset)                              # Reset the weights
    optimizer = optim.SGD(model.parameters(), lr=lr, momentum=momentum) # We choose SGD optimizer with lr=0.01
    for num_epoch in range(3): # 3 epochs to train model initially on trainset
        for batch_data in trainloader:
            # batch_data is a list of images and its labels
            inputs, labels = batch_data                                # seperate images from the labels
            inputs,labels= inputs.to(device), labels.to(device)  # convert to CUDA if available
            inputs=inputs.to(torch.float32)
            labels=labels.squeeze().to(torch.long)               # make labels compatible to our criterion
            optimizer.zero_grad()                                # reset the parameter gradients
            outputs = model(inputs)                              # forward pass
            
            loss = criterion(outputs, labels)                    # compute the loss to be minimised
            loss.backward()                                      # backward
            optimizer.step()                                     # update the gradient
        
    with torch.no_grad():                                        # testing phase, no need to compute the gradients
        running_loss=0
        k=0
        for batch_data in valloader_for_eval:                  # we use the valloader for evaluation which we created above
            img, label= batch_data
            inputs,label= img.to(device), label.to(device)
            inputs=inputs.to(torch.float32)
            pred_val=model(inputs)
            label=label.squeeze()
            label=label.to(torch.long)
            loss = criterion(pred_val, label)
            running_loss =running_loss + loss.item()
            k+=1
            
        loss_dict['my'+(str(model)[0:7])]=running_loss                # append loss to the dictionary with the model name
        
        print('For model {}, loss over validation set is {:.2f}'.format(model ,running_loss/k))
        print('================================================================')
```

```python
# we sort the values of dictionary and use the model name which has the lowest loss. we rename it best_model_A
sorted_dict=dict(sorted(loss_dict.items(), key=lambda item: item[1], reverse=False))

best_model=list(sorted_dict.keys())[0] # lowest loss
best_model_A= globals()[best_model]    # convert the string back to function
print("The best model as per the above loss calculation is {} with loss={:.2f} computed over the validation set" 
      .format(best_model,list(sorted_dict.values())[0]))
```

```python
# we now train a model using the best choice out of 4, on the overall train data i.e train+val set.
start3=time.time()                            # to calculate time taken for training
best_model_A.apply(parameter_weight_reset)    # reset weights and baises
for num_epoch in range(3): # 3 epochs
    running_loss, no_of_batches = 0, 0                    # to calculate the average loss across all batches
    for loader in [trainloader,valloader]:    # overall train data
        for batch_data in loader:
            optimizer = optim.SGD(best_model_A.parameters(), lr=lr, momentum=momentum)
            # batch_data is a list of images and its labels
            inputs, labels = batch_data
            inputs,labels= inputs.to(device), labels.to(device)
            inputs=inputs.to(torch.float32)
            labels=labels.squeeze().to(torch.long)               # make labels compatible to our criterion
            # reset the parameter gradients
            optimizer.zero_grad()
            # forward + backward + step
            outputs = best_model_A(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss = running_loss + loss.item()
            no_of_batches+=1
    print('after epoch {}, average loss is {}'.format(num_epoch+1,running_loss/no_of_batches))
end3=time.time()
```

```python
# we create two function, one to calculate the RMS error over the test set, and the other to compute the accuracy the
# accuracy score over the trained model for each dataset

def calc_RMSE_over_testset(model):
    with torch.no_grad():                           #testing phase, no need to compute the gradients
        for batch_data in testloader_for_eval:
            x,y= batch_data
            x,y= x.to('cpu'), y.to('cpu')             # we convert back to 'cpu' to use numpy operations
            x=x.to(torch.float32)
            y=y.to(torch.long)
            ypred=model.to('cpu')(x)                # ypred contains activation number for all 4 output neurons.
                                                    # The one with highest activation is the prediction of our network
            _, predicted_val = torch.max(ypred, 1)  # torch.max outputs the value and the index of neuron which
                                                    # has highest activation between the 4 output neurons
    return np.sqrt(mean_squared_error(y,predicted_val))  # calculate root mean square error

def accuracy_over_trained_model(dataset,model):
    with torch.no_grad():                           #testing phase, no need to compute the gradients
        model=model.to("cpu")
        total_samples, num_correct = 0, 0
        for batch_data in dataset:
            images, labels = batch_data
            images,labels= images.to('cpu'), labels.to('cpu')
            images=images.to(torch.float32)
            outputs = model(images)                 # output contains activation number for all 4 output neurons.
                                                    # The one with highest activation is the prediction of our network
            _, predicted = torch.max(outputs, 1)    # torch.max outputs the value and the index of neuron which
                                                    # has highest activation between the 4 output neurons
            total_samples+=labels.size(0)             # find the total number of labels in the dataset
            num_correct+=(predicted.reshape(-1,1) == labels).sum().item()         # find the number of correct ones
    return (100 * num_correct / total_samples)                  # computes the accuracy score 
```

We are now interested in computing the following metrics:

M1- Training accuracy: the prediction accuracy of a trained model on the training dataset.

M2- Validation accuracy: the prediction accuracy of a trained model on the validation dataset.

M3- Testing accuracy: the prediction accuracy of a trained model on the test dataset.

M4- Training time: the time taken to train the model (i.e. to learn/estimate the learnable parameters) on the training dataset.

```python
# We now calculate the metrics of interest
RMSE_model_A=calc_RMSE_over_testset(best_model_A)
M1_a=accuracy_over_trained_model(trainloader,best_model_A) # percent accuracy of trained model over the trainset
M2_a=accuracy_over_trained_model(valloader,best_model_A)   # percent accuracy of trained model over the valset
M3_a=accuracy_over_trained_model(testloader,best_model_A)  # percent accuracy of trained model over the testset
M4_a=end3-start3                                         # time taken to train the 1st model/parameters

print("""The accuracy percent of each metric from best CNN model having 2 conv and 2 linear layers is, 
M1={:.1f} % , M2={:.1f} % , M3={:.1f} % , M4={:.1f} seconds and the RMSE over test set={:.2f}"""
      .format(M1_a,M2_a,M3_a,M4_a,RMSE_model_A))
```

### C2(b) Three Conv layers and three FC layers

We now create 3 classes and 3 objects each having Three Conv layers and three FC layers

```python
class CNN_B_1(nn.Module):
    def __init__(self):
        super(CNN_B_1, self).__init__()
        self.conv1 = nn.Conv2d(1, 16, 7, padding=3) #1: #input channels; 16: #output channels; 7: kernel size; 3: padding
        self.conv2 = nn.Conv2d(16, 128, 6, padding=1)
        self.conv3 = nn.Conv2d(128, 32, 3, padding=1)
        self.fc1 = nn.Linear(32 * 4 * 4, 100)
        self.fc2 = nn.Linear(100, 20)
        self.fc3 = nn.Linear(20, 4)
        self.pool1 = nn.MaxPool2d(2, 2)
        self.pool2 = nn.MaxPool2d(4, 1)

    def forward(self, x):
        x = self.pool1(F.relu(self.conv1(x)))
        x = self.pool2(F.relu(self.conv2(x)))
        x = self.pool1(F.relu(self.conv3(x)))
        x = x.view(-1, 32 * 4 * 4)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x

myCNN_B_1 = CNN_B_1().to(device)          # convert to CUDA if available
summary(myCNN_B_1, input_size=(1,28,28))  # summary of the model created above
```

```python
class CNN_B_2(nn.Module):
    def __init__(self):
        super(CNN_B_2, self).__init__()
        self.conv1 = nn.Conv2d(1, 4, 5, padding=1) #3: #input channels; 4: #output channels; 5: kernel size; 2: padding
        self.conv2 = nn.Conv2d(4, 16, 5)
        self.conv3 = nn.Conv2d(16, 8, 4)
        self.pool1 = nn.MaxPool2d(5, 1)
        self.fc1 = nn.Linear(8 * 7 * 7, 200)
        self.fc2 = nn.Linear(200, 20)
        self.fc3 = nn.Linear(20, 4)

    def forward(self, x):
        x = self.pool1(F.selu(self.conv1(x)))
        x = self.pool1(torch.tanh(self.conv2(x)))
        x = self.pool1(F.relu(self.conv3(x)))
        x = x.view(-1, 8 * 7 * 7)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x

myCNN_B_2 = CNN_B_2().to(device)          # convert to CUDA if available
summary(myCNN_B_2, input_size=(1,28,28))  # summary of the model created above
```

```python
class CNN_B_3(nn.Module):
    def __init__(self):
        super(CNN_B_3, self).__init__()
        self.conv1 = nn.Conv2d(1, 4, 8, padding=3) #3: #input channels; 4: #output channels; 5: kernel size; 2: padding
        self.conv2 = nn.Conv2d(4, 8, 6, padding=1)
        self.conv3 = nn.Conv2d(8, 5, 5)
        self.fc1 = nn.Linear(5 * 20 * 20, 200)
        self.fc2 = nn.Linear(200, 40)
        self.fc3 = nn.Linear(40, 4)

    def forward(self, x):
        x = F.relu6(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = F.relu(self.conv3(x))
        x = x.view(-1, 5 * 20 * 20)
        x = torch.tanh(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = torch.sigmoid(self.fc3(x))
        return x

myCNN_B_3 = CNN_B_3().to(device)          # convert to CUDA if available
summary(myCNN_B_3, input_size=(1,28,28))  # summary of the model created above
```

```python
loss_dict2=dict()  # We create a dictionary to store the model name and the losses on validation set using that model
for model in [myCNN_B_1, myCNN_B_2, myCNN_B_3]:
    model.apply(parameter_weight_reset)                               # Reset the weights
    optimizer = optim.SGD(model.parameters(),lr=lr, momentum=momentum)
    for num_epoch in range(3): # 3 epochs
        for batch_data in trainloader:
            # batch_data is a list of images and its labels
            inputs, labels = batch_data                                     # seperate images from the labels
            inputs,labels= inputs.to(device), labels.to(device)       # to CUDA if available
            inputs=inputs.to(torch.float32)
            # reset the parameter gradients
            optimizer.zero_grad()
            # forward+backward+step
            outputs = model(inputs)
            labels=labels.squeeze().to(torch.long)               # make labels compatible to our criterion
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
        
    with torch.no_grad():                                              # testing phase, no need to compute the gradients
        running_loss, no_batch = 0, 0
        for batch_data in valloader_for_eval:
            img, label=batch_data
            inputs,label= img.to(device), label.to(device)
            inputs=inputs.to(torch.float32)
            pred_val=model(inputs)
            label=label.squeeze()
            label=label.to(torch.long)
            loss = criterion(pred_val, label)
            running_loss =running_loss + loss.item()
            no_batch+=1
        loss_dict2['my'+(str(model)[0:7])]=running_loss                    # append loss to the dictionary on the model name
        print('For model {}, loss over validation set is {:.2f}'.format(model ,running_loss/no_batch))
        print('================================================================')
```

```python
# sort dictionary to find lowest loss and the corresponding model name
sorted_dict2=dict(sorted(loss_dict2.items(), key=lambda item: item[1], reverse=False))
best_model_2=list(sorted_dict2.keys())[0]
best_model_B= globals()[best_model_2]                        # convert string back to the function name
print("The best model as per the above loss calculation is {} with loss={:.2f} computed over the validation set" 
      .format(best_model_2,list(sorted_dict2.values())[0]))
```

```python
# We now train the best model from above on the whole training set i.e train+val data
start4=time.time()
best_model_B.apply(parameter_weight_reset) # reset parameters
for num_epoch in range(3): # 3 epochs
    running_loss, batches = 0, 0
    for loader in [trainloader,valloader]:
        for batch_data in loader:
            optimizer = optim.SGD(best_model_B.parameters(),lr=lr, momentum=momentum)
            # batch_data is a list of images and its labels
            inputs, labels = batch_data
            inputs,labels= inputs.to(device), labels.to(device)
            inputs=inputs.to(torch.float32)
            labels=labels.squeeze().to(torch.long)               # make labels compatible to our criterion
            # reset the parameter gradients
            optimizer.zero_grad()
            # forward + backward + optimize
            outputs = best_model_B(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss = running_loss + loss.item()
            batches+=1
    print('after epoch {}, average loss is {}'.format(num_epoch+1,running_loss/batches))
end4=time.time()
```

```python
# We now calculate the metrics of interest
RMSE_model_B=calc_RMSE_over_testset(best_model_B)        
M1_b=accuracy_over_trained_model(trainloader,best_model_B) # percent accuracy of trained model over the trainset
M2_b=accuracy_over_trained_model(valloader,best_model_B)   # percent accuracy of trained model over the valset
M3_b=accuracy_over_trained_model(testloader,best_model_B)  # percent accuracy of trained model over the testset
M4_b=end4-start4                                # time taken to train the 2nd model/parameters
print("""The accuracy percent of each metric from best CNN model having 3 conv and 3 linear layers is, 
M1={:.1f} % , M2={:.1f} % , M3={:.1f} % , M4={:.1f} seconds and the RMSE over test set={:.2f}"""
      .format(M1_b,M2_b,M3_b,M4_b,RMSE_model_B))
```

# D. Unsupervised learning on Fashion-MNIST

```python
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize(mean=[0.5], std=[0.5])])

#Load the test data as main data as per instruction
dataset = datasets.FashionMNIST(root='./data', train=False, download=True, transform=transform)

classes = dataset.classes # check the class labels and names
print(dataset)
print('Classes of this dataset are:', classes)
```

We now choose two classes out of the 10 to do Unsupervised learning.
In this case, we chose the class named pullover(class id=2) and class named dress(class id=3)

```python
pullover_dataset, y_pullover, dress_dataset, y_dress =[], [], [], []

for i, data in enumerate(dataset, 0):
    images,labels=data
    if labels==2: #pullover class is labelled as 2
        pullover_dataset.append(images)
        y_pullover.append(labels)
    if labels==3: #dress class is labelled as 3
        dress_dataset.append(images)
        y_dress.append(labels)

pullover_dataset=torch.stack(pullover_dataset)               # We convert the image list to torch tensor
dress_dataset=torch.stack(dress_dataset)

y_pullover=np.array(y_pullover).reshape(-1,1)                # We convert the label list to numpy array
y_dress=np.array(y_dress).reshape(-1,1)

Unlabelled_data=torch.cat(((pullover_dataset,dress_dataset))) # We join the two class images and name it Unlabelled data
y=torch.cat((torch.tensor(y_pullover),torch.tensor(y_dress))) # Joining the labels of 2 classes
```

## D1- Dimensionality reduction and clustering
### D1(a)- Apply PCA to visualise top 5 PCs

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# We flatten our images so as to apply PCA
X_unlabelled = np.array(Unlabelled_data[:][0].numpy()).reshape(1, 28*28)

for i in range(1, len(Unlabelled_data[:])):
    X_unlabelled = np.append(X_unlabelled,np.array(Unlabelled_data[i][0].numpy()).reshape(1, 28*28),axis = 0)

# We scale our features before applying PCA
X_unlabelled = StandardScaler().fit_transform(X_unlabelled)

# Now we apply PCA and we chose the top 30 components out of 784 from original space(28*28)
pca = PCA(n_components=30)

# We apply PCA to our scaled data and we get transformed pca dataset with N*30 shape
X_pca_data = pca.fit_transform(X_unlabelled)
```

```python
# We now visualise the top 5 Eigenvectors as images and show the percentage of variance it covers
for i in range(5):
    plt.figure(figsize=(4, 3))
    plt.imshow(pca.components_[i].reshape(28,28),cmap = gray_scale_cmap)        # Reshape back to original form of 28*28
    plt.axis('off')
    plt.show()
    print('Image for Eigenvector {} and its overall variance contribution percent is {:.1f}%'
          .format(i+1,pca.explained_variance_ratio_[i]*100))
```

### D1(b)- Reconstruct images using 30 PCs

```python
# As we used 30 PCs to get transformed data, we can use inverse transform to roll it back to the original space of N*784
X_reconstructed = pca.inverse_transform(X_pca_data)

# first 1000 datapoints are those of pullover
plt.figure(figsize=(6, 3))
for i in range(5):
    plt.subplot(2,5,i+1)
    a=X_reconstructed[i].reshape(28,28)                                # Plotting from the reconstructed data
    plt.imshow(a,cmap = gray_scale_cmap)
    plt.axis('off')
    plt.subplot(2,5,i+6)
    plt.imshow(Unlabelled_data[i].reshape(28,28),cmap = gray_scale_cmap)   # Plotting from the original data
    plt.axis('off')
plt.show()
print('Reconstructed vs Original images for pullover')
print('=============================================')

# the last 1000 datapoints are those of dress
plt.figure(figsize=(6, 3))
for i in range(5):
    plt.subplot(2,5,i+1)
    a=X_reconstructed[i+1000].reshape(28,28)                              # Plotting using reconstructed data
    plt.imshow(a,cmap = gray_scale_cmap)
    plt.axis('off')
    plt.subplot(2,5,i+6)
    plt.imshow(Unlabelled_data[i+1000].reshape(28,28),cmap = gray_scale_cmap) # Plotting using original data
    plt.axis('off')
plt.show()
print('Reconstructed vs Original images for dress')
```

```python
# We choose the first two Principal Components
PC_1= X_pca_data[:,0].reshape(-1,1)
PC_2= X_pca_data[:,1].reshape(-1,1)

# We combine the 2 PCs to make a single dataset of shape 2000 * 2
top_2_PCs= np.column_stack((PC_1,PC_2))
```

### D1(c)- Visualise the two-dimensional PCA representations of all data points in a 2D plane

```python
# We plot a scatter plot of the 2 PC components against it original label
plt.figure(figsize=(9, 7))
plt.scatter(PC_1, PC_2, c=y, edgecolor='pink', 
            alpha=0.7, cmap=plt.cm.get_cmap('brg', 2))

print('The points in blue, labelled-2 are those of pullover, and the points in green, labelled-3 are those of dress')

plt.xlabel('PC_1')
plt.ylabel('PC_2')
plt.title('Two-dimensional PCA representations of all data points')
plt.legend(np.unique(y))
plt.colorbar();
plt.show();
```

### D1(d)- Spectral clustering of cluster all data points as represented by the top two PCs

### D1(e)- Autoencoder with five Conv2d layers and five ConvTranspose2d layers

```python
# We now create an Autoencoder class containing 5 Conv2D layers for Encoder and 5 ConvTranspose2d layers for Decoder
# We use ReLu(), Tanh() and Sigmoid() as our Activation functions
class Autoencoder(nn.Module):
    def __init__(self):
        super(Autoencoder, self).__init__() 
        self.encoder = nn.Sequential(      # Encoder
            
            nn.Conv2d(1, 5, 3, stride=2, padding=1), # 1st conv layer with 1 input image channel, 5 output channels, 3x3 square convolution, 1 padding
            nn.ReLU(),                               # Rectified Linear Unit
            nn.Conv2d(5, 9, 3, stride=2, padding=1), # 2nd conv layer with 5 input image channels, 9 output channels, 3x3 square convolution, 1 padding
            nn.ReLU(),
            nn.Conv2d(9, 16, 3, padding=2),          # 3rd conv layer with 9 input image channels, 16 output channels, 4x4 square convolution, 2 padding
            nn.ReLU(),
            nn.Conv2d(16, 25, 5),                    #4th conv layer with 16 input image channels, 25 output channels, 5x5 square convolution
            nn.Tanh(),                               # Hyperbolic Tangent
            nn.Conv2d(25, 40, 7, padding=2)          #5th conv layer with 25 input image channels, 40 output channels, 7x7 square convolution, 2 padding
        )
        
        self.decoder = nn.Sequential(        # Decoder
            nn.ConvTranspose2d(40, 25, 7, padding=2), #1st ConvTranspose2d layer with 40 input channels, 25 output channels, 7x7 convolution, 2 padding
            nn.Tanh(),
            nn.ConvTranspose2d(25, 16, 5),            #2nd ConvTranspose2d layer with 25 input channels, 16 output channels, 5x5 convolution
            nn.ReLU(),
            nn.ConvTranspose2d(16, 9, 3, padding=2),  #3rd ConvTranspose2d layer with 16 input channels, 9 output channels, 4x4 convolution, 2 padding
            nn.ReLU(),
            nn.ConvTranspose2d(9, 5, 3, stride=2, padding=1, output_padding=1),      #4th ConvTranspose2d layer with 9 input channels, 5 output channels
            nn.ReLU(),                                                               # 2 stride, 1 padding, 3x3 convolution
            nn.ConvTranspose2d(5, 1, 3, stride=2, padding=1, output_padding=1),      #5th ConvTranspose2d layer with 5 input channels, 1 output channels
                                                                                     # 2 stride, 1 padding, 3x3 convolution
            nn.Sigmoid()                      #to range [0, 1]
        )

    def forward(self, x):    # Forward pass
        x = self.encoder(x)
        x = self.decoder(x)
        return x

# Create an object for our class
my_auto_encoder=Autoencoder()
```

```python
#Training (optimisation) parameters
learning_rate=0.5
max_epochs = 20
batch_size=500
weight_decay=1e-5

# We choose mean square error loss
criterion = nn.MSELoss() 

# we choose the SGD optimiser
optimizer = torch.optim.SGD(my_auto_encoder.parameters(), lr=learning_rate, weight_decay=weight_decay)

#Specify how the data will be loaded in batches (with random shuffling)
# We use the unlabelled data of our 2 classes
data_loader = torch.utils.data.DataLoader(Unlabelled_data, batch_size=int(batch_size), shuffle=True)

#Start training

avg_loss_list=[]
for epoch in range(max_epochs):
    avg_loss, num_batches = 0, 0
    for batch_data in data_loader:
        image= batch_data
        optimizer.zero_grad()       # reset the gradients
                                    # forward + backward + step
        reconstructed = my_auto_encoder(image)
        loss = criterion(reconstructed, image)
        loss.backward()
        optimizer.step()
        avg_loss+=loss.item()
        num_batches+=1
    avg_loss_list.append(avg_loss/num_batches)
    if (epoch % 3) == 0:
        print(' After {} epochs, the loss is :{:.3f}'.format(epoch+1, float(avg_loss/num_batches)))
```

```python
plt.figure(figsize=(7, 5))
plt.plot(np.linspace(1,20,num=20),avg_loss_list)
plt.title('Loss against epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show();
```

## D2- Observations
#### 1.
We can observe that the images reconstructed using the 30 PCs contain traces of the other classes even after we roll it back to the original space. In our case, the images of pullovers contain some ghost images of dress and in reconstructed images of dress, we can see some ghosting and traces of pullovers. Also, in the Eigenvectors represented as images, one can see images of both the classes on top of each other. It is because the first few Eigenvectors cover the most variance of all data for both the classes. The combined variance our first 5 Eigenvectors can explain is around 56%.

#### 2.
In our case, where pullover and dress were chosen, the KMeans segregated the two clusters very wrongly, while the Spectral Clustering did a lot better. Eventhough Spectral Clustering did better at seperating the points into 2 classes with a fine cut, the number of points in each cluster was not same. The algorithm gave around 11% samples to the wrong cluster but still though, the majority were classified correctly as we saw in plot with reference to labels and this is really helpful in clustering unlabelled data.

#### 3.
In the loss vs epoch graph, the loss starts to steady down after 8 odd epochs. It doesn't improve as drastically as it had increased previously. If we train for even more epochs, the loss might decrease further and thus cause overfitting. We can control this overfitting by using l2 regularization parameter in our optimizer. It is represented by the weight_decay term. If this term was set to zero, our model might start to overfit. Increasing the learning rate a bit more further or increasing the batch size, might cause the gradient to skip the minimum and thus, resulting in more loss and graph jumping around. If the learning rate was low, it would take the model a large number of epochs to train and reduce loss. So, we have to decide properly on what hyperparameters to choose before training the AutoEncoder.
